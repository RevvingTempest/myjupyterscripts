## A Taste of Matplotlib

In this notebook we'll use the *pyplot* interface,  matplotlib's version of MATLAB. It is stateful. In particular, any plotting commands apply the the current window. Here, windows will be rendered inline in the notebook.

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

### Line Graphs

A linearly spaced array of values from -pi to pi.

````python
PI = np.pi
x = np.linspace(-PI, PI, 256)
````

````python
C, S = np.cos(x), np.sin(x)
````

The *plt* interface is matplotlib's version of MATLAB. It is stateful. In particular, any *plt* plotting command creates a new window and further *plt* commands apply to that window. In this case, windows are rendered inline in the notebook.

````python
plt.plot(x, C, color='green', linewidth=2.5, label='cosine')
plt.plot(x, S, color='red', linewidth=2.5, label='sine')
plt.xticks([-PI, -PI/2, 0, PI/2, PI], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1])
plt.legend(loc='upper left')
````

````python
plt.plot(x, np.sin(x), color='green')
plt.plot(x, np.sin(x + (PI/3)), color='blue', linestyle='--')
plt.plot(x, np.sin(x + (2*PI/3)), color=(1.0,0.5, 0.2))
plt.xlabel('Hello', color='red')
plt.ylabel('World')
plt.title('Pretty Cool Stuff')
````

### Bar Graphs

````python
n = 12
xs = np.arange(n)
ys = (1 - xs/float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(xs, ys, align='center', facecolor='blue', edgecolor='red')
plt.xlim(-0.5, 11.5)
plt.xticks(np.arange(12))
for x,y in zip(xs, ys):
    plt.text(x, y+0.04, '%.2f' %y, ha='center', va='center')
plt.ylim(0, 1.1)
````

### Histograms

Histograms give us a way to visualize the frequency distribution of data. hist returns the tuple (n, bins, patches), where n are the bin's values, bins are the locations of the bins, and patches are the artists that comprise the image. The argument *normed* is True if counts are normalized to form a distribution (sum to 1.0). *bins* is the number of bins.

````python
data = np.random.randn(4000)
plt.hist(data, bins=100, edgecolor='b')
````

````python
# The histtype argument controls the type of histogram:
# ‘bar’ is a traditional bar-type histogram. If multiple data are given the bars are aranged side by side.
# ‘barstacked’ is a bar-type histogram where multiple data are stacked on top of each other.
# ‘step’ generates a lineplot that is by default unfilled.
# ‘stepfilled’ generates a lineplot that is by default filled.
plt.hist(data, bins=40, color='pink', histtype='stepfilled', edgecolor='k')
````

If we set the *alpha* argument for transparency, we can overwrite multiple histograms on the same plot.

````python
ds = []
ds.append(np.random.normal(0, .8, 1000))
ds.append(np.random.normal(-2, 1, 1000))
ds.append(np.random.normal(3, 2, 1000))
colors = ['red', 'green', 'blue']
kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)
for i in range(3):
    plt.hist(ds[i], color=colors[i], **kwargs)
````

Matplotlib also supports 2D histograms where the data points are dropped into the cells of a grid.

Numpy's random.multivariate_normal function generalizes the normal distribution to higher dimensions. These are the two required arguments:
- mean : 1-D array_like, of length N
  - Mean of the N-dimensional distribution.
- cov : 2-D array_like, of shape (N, N)
  - Covariance matrix of the distribution. It must be symmetric and positive-semidefinite for proper sampling. *cov\[i\] \[j\]* is the covariance between the *i*'th and *j*'th random variable.

````python
N = 10000
mean = [0, 0]
cov = [[10, 10], [10, 20]]
x, y = np.random.multivariate_normal(mean, cov, N).T   # transpose to shape (2, N)
plt.hist2d(x, y, bins=30, normed=True, cmap='Greens', range=[[-10, 10], [-10, 10]])
cb = plt.colorbar()
cb.set_label('counts')
````

````python
plt.hexbin(x, y, gridsize=20, cmap='Blues', extent=(-10, 10, -10, 10))
cb = plt.colorbar()
cb.set_label('counts')
````

### Contour plots
Contour plots show the contours of a function from $R^2$ to $R$. They may be contour lines (using ```contour```) or filled contours (```countourf```). The *levels* argument is used to indicate the levels. If *levels* is an integer $n$, then $n+1$ contours lines are drawn (for $n$ data levels). If *levels* is an array of increasing values, contour lines are drawn at those values.

````python
def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
def g(x,y): return x**2 + y**2
def h1(x,y): return x*y**3 - y*x**3
def h2(x,y): return np.cos(np.abs(x)+np.abs(y))
def h3(x,y): return -x*y*np.e**(x**2-y**2)
n = 256
x = np.linspace(-5,5,n)
y = np.linspace(-5,5,n)
X,Y = np.meshgrid(x,y)

levels = 12

plt.contourf(X, Y, h1(X,Y), levels, alpha=.75, cmap=cm.cool)
plt.contour(X, Y, h1(X,Y), levels, colors='black', linewidth=.5)
````
